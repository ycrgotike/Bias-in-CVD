```python
# define numerical & categorical columns
categorical_features = [feature for feature in df.columns if df[feature].dtype == "O"]
numerical_features = [feature for feature in df.columns if df[feature].dtype != "O" and feature != "HeartDisease"]

# print columns
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
print('We have {} numerical features : {}'.format(len(numerical_features), numerical_features))
```

#### 3.8 Replace 0 with NaN

```python

df['RestingBP'] = df['RestingBP'].replace(0, np.nan)
df['Cholesterol'] = df['Cholesterol'].replace(0, np.nan)
```

#### 3.9 Filling NaN values with median based on gender of the patient


```python
df['RestingBP'] = df.groupby('Sex')['RestingBP'].transform(lambda x: x.fillna(x.median()))
df['Cholesterol'] = df.groupby('Sex')['Cholesterol'].transform(lambda x: x.fillna(x.median()))
```

#### Bivariate Analysis: Numerical × Target

```python
import math
n_cols = 3
n_rows = math.ceil(len(numerical_features) / n_cols)
plt.figure(figsize=(18, 5 * n_rows))


for i, col in enumerate(numerical_features, 1):
    plt.subplot(n_rows, n_cols, i)
    
    # Histogram + KDE for No Disease
    sns.histplot(
        df[df['HeartDisease'] == 0][col],
        kde=True,
        alpha=0.4,
        label='No Heart Disease'
    )
    
    # Histogram + KDE for Disease
    sns.histplot(
        df[df['HeartDisease'] == 1][col],
        kde=True,
        alpha=0.4,
        label='Heart Disease'
    )
    
    plt.title(f"{col} Distribution by HeartDisease")
    plt.legend()

plt.tight_layout()
plt.show()
```

##### Insights
- Age - Patients with heart disease tend to be older, with the disease distribution shifted toward higher age ranges compared to those without disease.
- RestingBP distributions overlap considerably between groups, indicating limited standalone predictive strength.
- Cholesterol levels show heavy overlap between disease and non-disease groups, suggesting weak discriminatory power on their own.
- FastingBS - Elevated fasting blood sugar (1) appears more frequently in the heart disease group, indicating a positive association with disease presence.
- MaxHR - Individuals with heart disease generally exhibit lower maximum heart rates, showing a clear leftward shift compared to non-diseased individuals.
- Higher Oldpeak values are predominantly observed in heart disease patients, demonstrating strong separation and high predictive potential.
